```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
plt.style.use('fivethirtyeight')
mpl.rcParams['lines.linewidth'] = 2
```

```python
year = list(range(1776, 2020, 1))
```

```python
president = [0] * len(year)
president[-4:] = [1] * 4
```

```python
plt.figure(figsize=(10, 7))
plt.plot(year, president)
plt.title('Number of US Presidents Who Are Stupid Idiots')
plt.yticks([0, 1])
plt.xticks([1776, 1800, 1850, 1900, 1950, 2000, 2020])
plt.show()
```
